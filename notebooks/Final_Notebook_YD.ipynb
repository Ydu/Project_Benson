```python
# Get the necessary imports 
import os
import re
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import date
import calendar

# plot modules
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
matplotlib.style.use("ggplot")
```

```python
# Consolidate data from multiple csv files into one data frame
# In this case, all data from 3 months before the beginning of this summer
# Start by getting csv files into their own DataFrames
new_cols = ["C/A", "UNIT", "SCP", "STATION", "LINENAME", "DIVISION", "DATE", "TIME", "DESC", "ENTRIES", "EXITS"]

mar02_2019_path = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_190302.txt"
mar02_2019_mta_data = pd.read_csv(mar02_2019_path, sep=",", header=0, names=new_cols)

mar09_2019_path = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_190309.txt"
mar09_2019_mta_data = pd.read_csv(mar09_2019_path, sep=",", header=0, names=new_cols)

mar16_2019_path = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_190316.txt"
mar16_2019_mta_data = pd.read_csv(mar16_2019_path, sep=",", header=0, names=new_cols)

mar23_2019_path = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_190323.txt"
mar23_2019_mta_data = pd.read_csv(mar23_2019_path, sep=",", header=0, names=new_cols)

mar30_2019_path = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_190330.txt"
mar30_2019_mta_data = pd.read_csv(mar30_2019_path, sep=",", header=0, names=new_cols)

apr06_2019_path = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_190406.txt"
apr06_2019_mta_data = pd.read_csv(apr06_2019_path, sep=",", header=0, names=new_cols)

apr13_2019_path = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_190413.txt"
apr13_2019_mta_data = pd.read_csv(apr13_2019_path, sep=",", header=0, names=new_cols)

apr20_2019_path = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_190420.txt"
apr20_2019_mta_data = pd.read_csv(apr20_2019_path, sep=",", header=0, names=new_cols)

apr27_2019_path = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_190427.txt"
apr27_2019_mta_data = pd.read_csv(apr27_2019_path, sep=",", header=0, names=new_cols)

may04_2019_path = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_190504.txt"
may04_2019_mta_data = pd.read_csv(may04_2019_path, sep=",", header=0, names=new_cols)

may11_2019_path = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_190511.txt"
may11_2019_mta_data = pd.read_csv(may11_2019_path, sep=",", header=0, names=new_cols)

may18_2019_path = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_190518.txt"
may18_2019_mta_data = pd.read_csv(may18_2019_path, sep=",", header=0, names=new_cols)

may25_2019_path = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_190525.txt"
may25_2019_mta_data = pd.read_csv(may25_2019_path, sep=",", header=0, names=new_cols)
```

```python
# Concatenate all new DataFrames into one DataFrame
mta_datasets_2019 = pd.concat([mar02_2019_mta_data, mar09_2019_mta_data, mar16_2019_mta_data,
mar23_2019_mta_data, mar30_2019_mta_data, apr06_2019_mta_data, apr13_2019_mta_data, apr20_2019_mta_data,
apr27_2019_mta_data, may04_2019_mta_data, may11_2019_mta_data, may18_2019_mta_data, may25_2019_mta_data])
```

```python
# Add a column for total turnstile passes
# and another column for number of passes in each 4 hour block
mta_datasets_2019['TURNSTILE_PASSES'] = mta_datasets_2019['ENTRIES'] + mta_datasets_2019['EXITS']
#mta_datasets_2019['PASSES_BY_HOUR'] = mta_datasets_2019.TURNSTILE_PASSES.diff()
```

```python
station_datetime = ["STATION", "C/A", "UNIT", "SCP", "DATE", "TIME"]
mta_datasets_2019_grouped_by_station = mta_datasets_2019.set_index(station_datetime).sort_index()
```

```python
mta_datasets_2019_grouped_by_station = mta_datasets_2019_grouped_by_station.reset_index()
mta_datasets_2019_grouped_by_station['PASSES_BY_HOUR'] = mta_datasets_2019_grouped_by_station['TURNSTILE_PASSES'].diff()
mta_datasets_2019_grouped_by_station
```

```python
mta_datasets_2019_grouped_by_station[mta_datasets_2019_grouped_by_station.TIME == '19:00:00'].sort_values(by='PASSES_BY_HOUR')
```

```python
from collections import defaultdict
mean_traffic_5pm = defaultdict(int)

for station in mta_datasets_2019_grouped_by_station.STATION.unique():
    mean_traffic_5pm[station] = np.mean(mta_datasets_2019_grouped_by_station[(mta_datasets_2019_grouped_by_station.STATION == station) &
                                                                             ((mta_datasets_2019_grouped_by_station.TIME == '19:00:00') |
                                                                              (mta_datasets_2019_grouped_by_station.TIME == '20:00:00') |
                                                                              (mta_datasets_2019_grouped_by_station.TIME == '21:00:00'))]
                                                                            .PASSES_BY_HOUR)
```
